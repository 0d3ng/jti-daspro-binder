# Student Jobsheet Usage Guide
This jobsheet will be used to carry out the Basic Programming Practicum at the State Polytechnic of Malang, Information Technology Department. The main component of this jobsheet is to use `Java Shell tool (JSheel)`, so it will be very easy
to learn more interactive programming without installing additional tools or editors.

## Access Jobsheet

There are 2 modes or views that can be used, namely classic mode (simple) and lab mode. Classic mode is synonymous with
simplicity of the appearance of the editor, but the essence of the tool to run program code can still be done. The opposite of
lab mode, like a tool for the lab with the complexity of the settings in it up to
editor view. Here are the views of the two modes

![](images/mode-classic.png)
> User Interface Classic mode

![](images/mode-lab.png)
> Lab Mode User Interface

The steps you can take to use this jobsheet are as follows
+ Please click the link [classic mode](https://s.id/binder-classic) to select the classic view, while clicking
link [lab mode](https://s.id/binder-lab) for lab mode display.
+ After clicking the link using either classic mode or lab mode, a loading screen will appear as shown in the following image

    ![](images/loading-lab.png)

+ Please wait a while until the display appears whether using classic mode or lab mode, select the `jobsheet` . folder
as desired. For example jobsheet 5 or week 5 means that the selected folder is `05`. After
selecting the folder will enter the directory as shown in the following image

  ![](images/01.png)

+ Reselect the file with the extension `.ipynb` as in the image below,

  ![](images/02.png)

  Then it will open a new tab, with the jobsheet view or editor ready to use. It looks like the following picture

  ![](images/03.png)

  From the picture above there are two modes, namely `Markdown and Code`. `Markdown` means contains description of the jobsheet or steps
  which must be done. Then for the `Code` mode, it contains program code that can be run directly. Here's the visualization
of the two modes.

  > `Markdown and Code` are often referred to as `Cell`, in fact apart from these two there are other types but
  > frequently used are `Markdown and Code.`

  ![](images/04.png)

  When clicked on `Code` mode then the top circled in red will change to `Code` mode, when
has changed to `Code` mode can click the `Run` button to run. For more details, see the illustration in
below

  ![](images/05.png)

  > Every `Cell`, whether it is `Code or Markdown` type, can be executed with the Run button or with a combination
  > `CTRL + Enter` key. When we want to change the value or content contained in `Cell`, we can do this by
  > `double click`, then when we want to add `Cell` below or above it can be found from the menu
  > `Insert - Insert Cell Above (insert above current Cell) or vice versa Insert Cell Bellow (below current Cell)`
  >
  > **Important Notes!**
  + Markdown is a lightweight markup language for formatting documents, for more details and how to use them
  can be seen at the link [below](https://www.markdownguide.org/getting-started/)
  + The editor you are using is on a public server that has a certain session length, so that when it time out
  then have to repeat from the beginning for the process of using this jobsheet.

## Troubleshooting
There are several things or cases that often occur when using this online jobsheet such as
+ When running the code but no output appears, if this happens then what needs to be done is to restart the kernel as shown in the following picture
  + Click Save until a checkpoint appears as shown in the following image

    ![](images/09.png)

  + Reopen the jobsheet file as shown below

    ![](images/10.png)

+ Session has run out or has been idle for too long, resulting in program code or anything that has been typed will be lost, the steps that can be taken are to repeat from the beginning or click the link for this jobsheet. The session that has been exhausted is shown in the image below
