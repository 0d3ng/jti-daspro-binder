# Panduan Penggunaan Jobsheet Mahasiswa
Jobsheet ini yang akan digunakan untuk melakukan praktikum Dasar Pemrograman Jurusan Teknologi Informasi Politeknik Negeri
Malang. Komponen utama dari jobsheet ini adalah menggunakan `Java Shell tool (JSheel)`, sehingga akan sangat memudahkan
untuk belajar pemrograman yang lebih interaktif tanpa installasi tool atau editor tambahan.

## Akses Jobsheet

Ada 2 mode atau tampilan yang bisa digunakan, yaitu mode klasik (simple) dan mode lab. Mode klasik adalah identik dengan
kesederhanaan tampilan editor, tetapi esensi dari tool untuk menjalankan kode program masih bisa dilakukan. Kebalikan dari
mode lab, layaknya sebuah tool untuk lab dengan kompleksitas dari pengaturan-pengaturan yang ada di dalamnya sampai dengan
tampilan editornya. Berikut ini tampilan-tampilan dari kedua mode tersebut

![](images/mode-classic.png)
> User Interface Mode classic

![](images/mode-lab.png)
> User Interface Mode Lab

Langka-langkah yang dapat anda lakukan untuk menggunakan jobsheet ini adalah sebagai berikut
+ Silakan klik link [mode classic](https://s.id/binder-classic) untuk memilih tampilan classic, sedangkan klik
link [mode lab](https://s.id/binder-lab) untuk tampilan mode lab.
+ Setelah klik link baik menggunakan mode classic atau mode lab, nanti akan muncul tampilan loading seperti pada gambar berikut

    ![](images/loading-lab.png)

+ Silakan tunggu beberapa saat sampai muncul tampilan baik itu menggunakan mode classic atau mode lab, pilih folder `jobsheet`
sesuai dengan yang diinginkan. Sebagai contoh jobsheet 5 atau minggu ke-5 bearti yang dipilih ada folder `05`. Setelah
memilih folder tersebut akan masuk ke dalam direktori seperti ditunjukkan pada gambar berikut

  ![](images/01.png)

+ Pilih kembali file dengan ekstensi `.ipynb` seperti pada gambar di bawah ini,

  ![](images/02.png)

  Kemudian akan membuka tab baru, dengan tampilan jobsheet atau editor yang siap untuk digunakan. Tampilannya seperti gambar berikut

  ![](images/03.png)

  Dari gambar di atas terdapat dua mode, yaitu `Markdown dan Code`. `Markdown` berarti berisi deskripsi dari jobsheet atau langkah-langkah
  yang harus dikerjakan. Kemudian untuk mode `Code`, berisi kode program yang bisa langsung dijalankan. Berikut visualisasi
dari kedua mode tersebut.

  > `Markdown dan Code` ini sering disebut dengan `Cell`, sebenarnya selain kedua tersebut terdapat jenis yang lain akan tetapi
  > yang sering digunakan adalah `Markdown dan Code.`

  ![](images/04.png)

  Ketika diklik pada mode `Code` maka pada bagian atas yang dilingkari merah akan berubah menjadi mode `Code`, ketika
sudah berubah menjadi mode `Code` bisa klik tombol `Run` untuk menjalankan. Untuk lebih jelasnya perhatikan ilustrasi di
bawah ini

  ![](images/05.png)

  > Setiap `Cell` baik itu jenisnya `Code ataupun Markdown`, bisa dijalankan yaitu dengan tombol Run ataupun dengan kombinasi
  > tombol `CTRL + Enter`. Ketika kita ingin mengubah nilai atau isi yang terdapat di dalam `Cell`, bisa dilaukan dengan cara
  > `double klik`, kemudian ketika kita ingin menambahkan `Cell` di bawahnya atau di atasnya dapat ditemukan dari menu
  > `Insert - Insert Cell Above (insert di atas current Cell) atau sebaliknya Insert Cell Bellow (di bawah current Cell)`
  >
  > **Catatan Penting!**
  + Markdown adalah sebuah bahasa markup yang ringan untuk melakukan format dokumen, untuk lebih jelasnya dan cara penggunaannya
  dapat dilihat pada link [berikut](https://www.markdownguide.org/getting-started/)
  + Editor yang sedang Anda gunakan berada di public server yang memiliki lama session tertentu, sehingga ketika time out
  maka harus mengulangi dari awal untuk proses penggunaan jobsheet ini.


## Troubleshooting
Ada beberapa hal atau kasus yang sering terjadi ketika menggunakan jobsheet online ini seperti
+ Ketika menjalankan code tetapi tidak muncul output, jika terjadi hal tersebut maka yang perlu dilakukan adalah bisa dengan cara restart kernel seperti pada gambar berikut
  + Klik Save sampai muncul checkpoint seperti ditunjukkan pada gambar berikut

    ![](images/09.png)

  + Buka kembali file jobsheetnya seperti gambar di bawah ini

    ![](images/10.png)

+ Session sudah habis atau terlalu lama idle sehingga mengakibatkan kode program atau apapun yang telah diketik maka akan hilang, langkah yang dapat dilakukan adalah dengan cara mengulang dari awal atau klik tautan untuk jobsheet ini. Session yang telah habis ditunjukkan pada gambar di bawah ini

  ![](images/08.png)

## Pengumpulan Tugas*

Untuk model tugas sesuai dengan kesepakatan dan masing-masing dosen seperti apa, berikut ini adalah contoh pengumpulan tugas
menggunakan github.

> Untuk menggunakan github, mahasiswa harus sudah mendaftar terlebih dahulu di [github.com](https://github.com/) dan
> melakukan installasi git client, sebagai contoh [git dekstop](https://desktop.github.com/). Sebaiknya
> untuk proses pendaftaran menggunakan email institusi (xxxx@polinema.ac.id) agar mendapatkan tawaran benefit dari github.
>
> Ketika Anda ingin mengetik Markdown silakan menggunakan editor yang menurut Anda paling mudah seperti [Vs Code](https://code.visualstudio.com/download) atau editor
> online seperti [stackedit.io](https://stackedit.io/app#).

Berikut ini langkah-langkah yang dapat digunakan untuk melakukan pengumpulan tugas;
+ Silakan buka tempat pengumpulan tugas di github, untuk alamatnya nanti akan diinfokan dosen masing-masing. Sebagai contoh
sudah diberikan alamat pengumpulan, kemudian klik tombol `Fork` di pojok kanan atas seperti pada tampilan berikut

    ![](images/06.png)

+ Jika sudah menambahkan konten atau isi laporan Anda sesuai dengan panduan, Push dan Pull request untuk mengumpulkan menggunakan
tool git client yang sudah Anda install
+ Hasilnya nanti akan muncul pada grup telegram seperti pada contoh berikut

    ![](images/07.png)
